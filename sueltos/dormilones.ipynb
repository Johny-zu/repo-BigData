```python
# Installing uszipcode library
!pip install uszipcode==0.2.6

# Import the library and create a search engine
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode=True) # set simple_zipcode=True to use the simple search engine

# Define a function to get the state
def get_state_from_zipcode(zip_code):
    zipcode_info = search.by_zipcode(zip_code)
    if zipcode_info:
        return zipcode_info.state
    else:
        return None
```

```python
def labeled_barplot(data, feature):
    plt.figure(figsize = (10, 5))
    plt.xticks(rotation = 90, fontsize = 15)
    ax = sns.countplot(data=data, x=feature, order=data.groupby([feature])[feature].count().sort_values().index)

    for i in range(len(ax.patches)):
        x = ax.patches[i].get_x() + ax.patches[i].get_width()/2 - 0.25
        y = ax.patches[i].get_height() + 0.2
        ax.annotate('{:.1f}%'.format(ax.patches[i].get_height()/data.shape[0]*100), (x, y))
```

```python
def histogram_boxplot(data, feature, kde = False, bins = 23):
    fig, ax = plt.subplots(2, 1, sharex = True, gridspec_kw = {"height_ratios": (1, 3)}, figsize=(12, 7))
    sns.boxplot(data = data, x = feature, ax = ax[0], showmeans = kde, color = "violet")
    sns.histplot(data = data, x = feature, kde = kde, ax = ax[1], bins = bins, palette = "winter")
    ax[1].axvline(data[feature].mean(), color = "green", linestyle = "--") # Add mean to the histogram
    ax[1].axvline(data[feature].median(), color = "black", linestyle = "-.") # Add median to the histogram
```

```python
histogram_boxplot(df_datos, 'rate', True, 20)
```

```python
histogram_boxplot(df_datos, 'age', True, 20)
```
